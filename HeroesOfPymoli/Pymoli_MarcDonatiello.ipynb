```python
#Top Spenders
#Identify the top 5 spenders in teh game by total purchase value, then list (in a table): SN, Purchase Count,
#Average Purchase Price, Total Purchase Value
#Purchase Count
purch_count_top_df = df_purchase_data[["SN", "Purchase ID"]].drop_duplicates()
purch_group_top = purch_count_top_df.groupby(["SN"])
summ_purchcounttop_df = purch_group_top.count()
summ_purchcounttop_df.rename(columns={'Purchase ID':'Purchase Count'}, inplace=True)
summ_purchcounttop_df = summ_purchcounttop_df.sort_values(["Purchase Count"], ascending=False)
summ_purchcounttop_df.index.name='Player Name'
summ_purchcounttop_df.head(10)
```

```python
#Top Spenders
#Identify the top 5 spenders in teh game by total purchase value, then list (in a table): SN, Purchase Count,
#Average Purchase Price, Total Purchase Value
#Average Purchase Price
avg_pricetop_df = df_purchase_data[["SN", "Price"]]
purch_groupprice_top = avg_pricetop_df.groupby(["SN"])
summ_avgpricetop_df = purch_groupprice_top.mean()
summ_avgpricetop_df = round(summ_avgpricetop_df, 2)
summ_avgpricetop_df = summ_avgpricetop_df.sort_values(["Price"], ascending=False)
summ_avgpricetop_df.rename(columns={'Price' : 'Average Purchase Price'}, inplace=True)
summ_avgpricetop_df = summ_avgpricetop_df.replace('\$','', regex=True).astype(float)
summ_avgpricetop_df.index.name='Player Name'
summ_avgpricetop_df.head(10)
```

```python
#Top Spenders
#Identify the top 5 spenders in teh game by total purchase value, then list (in a table): SN, Purchase Count,
#Average Purchase Price, Total Purchase Value
#Total Purchase Value
summ_purchvaluetop_df = df_purchase_data.groupby(['SN']).agg({'Price':'sum'})
summ_purchvaluetop_df = summ_purchvaluetop_df.sort_values(["Price"], ascending=False)
summ_purchvaluetop_df.rename(columns={'Price' : 'Total Purchase Value'}, inplace=True)
summ_purchvaluetop_df.sort_values(by=["Total Purchase Value"],ascending=False)
summ_purchvaluetop_df.index.name='Player Name'
summ_purchvaluetop_df.head(10)
```

```python
#Top Spenders dataframe by Total Purchase Value
topspenders_df_col = pd.concat([summ_purchcounttop_df, summ_avgpricetop_df, summ_purchvaluetop_df], axis=1)
topspenders_df_col = topspenders_df_col.sort_values(["Total Purchase Value"], ascending=False)
topspenders_df_col.head(5)
#print(topspenders_df_col.head(5))
```

```python
#Most popular items - purchase count
purch_count_pop_df = df_purchase_data[["Item ID", "Item Name", "Purchase ID"]].drop_duplicates()
purch_group_pop = purch_count_pop_df.groupby(['Item ID', 'Item Name'])
summ_purchcountpop_df = purch_group_pop.count()
summ_purchcountpop_df.rename(columns={'Purchase ID':'Purchase Count'}, inplace=True)
summ_purchcountpop_df.rename(columns={'Price':'Item Price'}, inplace=True)
summ_purchcountpop_df = summ_purchcountpop_df.sort_values(["Purchase Count"], ascending=False)
summ_purchcountpop_df.head(10)
```

```python
#Most popular items - total purchase value
summ_purchvaluepop_df = df_purchase_data[["Item ID", "Item Name", "Purchase ID", "Price"]]
summ_purchvaluepop_df = df_purchase_data.groupby(['Item ID', 'Item Name']).agg({'Price':'sum'})
summ_purchvaluepop_df = summ_purchvaluepop_df.sort_values(["Price"], ascending=False)
summ_purchcountpop_df.rename(columns={'Purchase ID':'Purchase Count'}, inplace=True)
summ_purchvaluepop_df.rename(columns={'Price' : 'Total Purchase Value'}, inplace=True)
summ_purchvaluepop_df.sort_values(by=["Total Purchase Value"],ascending=False)
summ_purchvaluepop_df.head(10)
```

```python
#Most popular items - Item Price
item_price_df = df_purchase_data[["Item ID", "Item Name", "Price"]]
item_price_df.rename(columns={'Price':'Item Price'}, inplace=True)
item_price_df.head(10)
```

```python
#Most popular items - merge purchase count and total purchase value
merge_table = pd.merge(summ_purchcountpop_df, summ_purchvaluepop_df, on=["Item ID", "Item Name"])
merge_table_df = merge_table
purchcountvalue_df = merge_table
purchcountvalue_df.head()
```

```python
#Most Popular Items by purchase count
mostpop_items_df = pd.merge(purchcountvalue_df, item_price_df, on=["Item ID", "Item Name"]).drop_duplicates()
mostpop_items_df.head(40)
```

```python
#Most Profitable Items by total purchase value
mostpop_items_df = pd.merge(purchcountvalue_df, item_price_df, on=["Item ID", "Item Name"]).drop_duplicates()
mostpop_items_df = mostpop_items_df.sort_values(["Total Purchase Value"], ascending=False)
mostpop_items_df.head(5)
```
